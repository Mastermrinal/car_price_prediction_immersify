```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler



# Selecting relevant columns for feature engineering
features = df[['age', 'annual Salary', 'credit card debt', 'net worth']]

# Initialize MinMaxScaler
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)

# Convert scaled features back to a DataFrame
scaled_features = pd.DataFrame(scaled_features, columns=['age', 'annual Salary', 'credit card debt', 'net worth'])

# Add interaction terms
scaled_features['age_salary_interaction'] = scaled_features['age'] * scaled_features['annual Salary']
scaled_features['debt_to_income_ratio'] = scaled_features['credit card debt'] / (scaled_features['annual Salary'] + 1e-6)
scaled_features['net_worth_to_income_ratio'] = scaled_features['net worth'] / (scaled_features['annual Salary'] + 1e-6)

# Include the target variable
scaled_features['car_purchase_amount'] = df['car purchase amount']
```

```python
##checking for staionarity
from statsmodels.tsa.stattools import adfuller
def ad_test(dataset):
  dftest=adfuller(dataset,autolag='AIC')
  print("1.ADF:",dftest[0])
  print("2.p-value:",dftest[1])
  print("3.number of lags:",dftest[2])
  print("4.number of observations:",dftest[3])
  print("5.critical values:")
  for key,val in dftest[4].items():
    print("\t",key,":",val)
```

```python
ad_test(scaled_features['car_purchase_amount'])
```

```python
model_predictions=[]
n_test_obser=len(test_data)
```

```python
for i in range(n_test_obser):
  model=ARIMA(train_data,order=(5,2,0))
  model_fit=model.fit()
  output=model_fit.forecast()
  yhat=list(output)[0]
  model_predictions.append(yhat)
  actual_test_value=test_data[i]
  train_data.append(actual_test_value)
```

```python
plt.figure(figsize=(20,6))
plt.grid(True)
data_range=df[to_row:].index
plt.plot(data_range,test_data,color='red',marker='o',linestyle='dashed',label='Actual_price')
plt.plot(data_range,model_predictions,color='blue',label='Predicted_price')
plt.title('Car Price Prediction')
plt.xlabel('car price')
plt.ylabel('annual income')
plt.legend()
plt.show()
```
